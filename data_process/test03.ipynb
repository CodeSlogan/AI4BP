```python
import pandas as pd
```

```python
data_path = './data/Cuff-Less Blood Pressure Estimation/Samples/rec_1.csv'
```

```python
data = pd.read_csv(data_path, header=None)
```

```python
print(data)
```

```python
ppg_arr = data.loc[0, :]
abp_arr = data.loc[1, :]
```

```python
print(ppg_arr)
print(abp_arr)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure()

plt.plot(ppg_arr.loc[:1024])

plt.show()
```

```python

plt.plot(abp_arr.loc[:1024], color='red')

plt.show()
```

```python
n_sequences = len(ppg_arr) // 1024
n_sequences
```

```python
import numpy as np
```

```python
ppg_matrix = np.empty((n_sequences, 1024), dtype=np.float64)
abp_matrix = np.empty((n_sequences, 1024), dtype=np.float64)

#   
for i in range(n_sequences):  
    start = i * 1024  
    end = start + 1024  
    ppg_matrix[i, :] = ppg_arr[start:end].values
    abp_matrix[i, :] = abp_arr[start:end].values
```

```python
print(ppg_matrix.shape)
print(ppg_matrix)
```

```python
print(abp_matrix.shape)
print(abp_matrix)
```
